## Household's Problem (HH)
### List of Notations:

* <u>Two periods</u>: Young ($y$) and Old ($o$)
* <u>Endogenous</u>: $C_y,~ C_o,~ S$
* <u>Exogenous</u>:
 * $\phi_y$: Survival Probability in the young period (*of the first-period Old or the parents generation*)
 * $\phi_o$: Survival Probability in the old period (*of the second-period Old or children generation*)
 * $\tau_y$: Income tax on the Young
 * $T_o$:    Government transfer to the second-period Old
 * $b_y$:    The average bequest **received** by young households in the first period
 * $b_o$:    The average bequest **given** by young households in the second period
 * $R$: Gross Real Interest Rate ($R = 1+r-\delta$)

## Household's Problem (HH)
### Assumptions (HH) (part 1):

* Static Population: $\bar{N}$

* 2-period (young and old), homogeneous, household
* CRRA utility form ($\sigma = 1$) for simplification
* PAYG tax system. During the young period:
 * ($\tau_y \phi_y$):   High survival probability in the young period ($\phi_y \uparrow$) $\implies$ tax pressure on the young ($\tau_y \phi_y \uparrow$)
 * ($\frac{T_o}{\phi_o}$): The old receives less Gov't transfer ($T_o \downarrow$) as the old-period suvivability increases ($\phi_o \uparrow$)
* **Receive and Give "Accidental Bequests"** ($b_y$ and $b_o$) due to **precautionary saving** in old age


### Assumptions (HH) (part 2):
##### Delayed Bequests and Aging:

$\frac{b_y}{\phi_y}$: Inverse relationship between $b_y$ and $\phi_y$.

$\frac{b_o}{\phi_o}$: Inverse relationship between $b_o$ and $\phi_o$.

 * The old decumulates assets as longevity rises (Life-Cycle assumption)
 * The young family members expend more resources to look after the old (e.g., *strategic bequest assumption*, *old age dependency ratio*,...)
 * Especially in countries where pension and public health services are of little to no help.


### Assumptions (HH) (part 3):
##### Return on Savings and Aging:

$\frac{RS}{\phi_o}$: Inverse relationship between $R$ and $\phi_o$.

Assuming **Static Population (and/or falling Birth rates)**, High old-age survivability in old period ($\phi_o \uparrow$) $\implies$: 
* low return on insurance
* low return on capital investment ($MP_k \downarrow$)
* more loanable fund supply $\implies$ lower real return
* Investment starting age and investment return

## Household Problem (HH)
### The Basic Model (HH):

$$\max_{c_y, c_o}\quad \log C_y + \beta \phi_o \log C_o $$

s.t.:

\begin{align} c_y + s &= w (1-\tau_y \phi_y) + \frac{b_y}{\phi_y} \label{eq:1}\tag{1} \\ c_o + \frac{b_o}{\phi_o} &= \frac{RS}{\phi_o} + \frac{T_o}{\phi_o} \label{eq:2}\tag{2} \end{align}

(1) & (2) $\implies$ *the following Household's Budget Constraint*:

$$c_y + \frac{c_o \phi_o}{R} + \frac{b_o}{R} = w(1-\tau_y \phi_y) + \frac{b_y}{\phi_y} + \frac{T_o}{R} \label{eq:3} \tag{3}$$


## Analytical Findings (HH)
### First Order Conditions (HH):

\begin{align} S &= \frac{\beta\phi_o}{1+\beta\phi_o}  \Big(w(1-\tau_y \phi_y) + \frac{b_y}{\phi_y}\Big) + \frac{1}{1+\beta \phi_o} \Big( \frac{b_o - T_o}{R} \Big) \label{eq:4}\tag{4} \\   
C_y &= \frac{1}{1+\beta\phi_o} \Big(w(1-\tau_y \phi_y) + \frac{b_y}{\phi_y} + \frac{T_o - b_o}{R}\Big) \label{eq:5} \tag{5}\\
C_o &= \frac{\beta R}{1+\beta \phi_o} \Big(w(1-\tau_y \phi_y) + \frac{b_y}{\phi_y} \Big) + \frac{\beta}{1+\beta\phi_o} (T_o - b_o) \label{eq:6}\tag{6} \end{align}


### Euler equation (HH):

$$ \frac{MU_{c_y}}{MU_{c_o}} = \frac{c_o}{c_y} = \beta R $$



```python
B, R, phi_o, phi_y, w, tax, T_o, b_y, b_o = symbols('B, R, phi_o, phi_y, w, tax, T_o, b_y, b_o')


S = ((B*phi_o/(1+B*phi_o)) * (w*(1-tax*phi_y) + b_y/phi_y)) + ((1/(1+B*phi_o)) * (b_o - T_o)/R)

dSdR = diff(S, R)
dSdphi_y = diff(S, phi_y)
dSdphi_o = diff(S, phi_o)
dSdtax = diff(S, tax)
init_printing()

print('Derivative of Saving w.r.t:') 

print('\n\n (R) Gross Interest Rate =')
display(dSdR)

print('\n\n (phi_y) Survival probability in the first period (of the parent generation) =')
display(dSdphi_y)  

print('\n\n (phi_o) Survival probability in the second period (of the children generation) =')
display(dSdphi_o)  

print('\n\n (tau_y) Income Tax =')
display(dSdtax)      

hide_toggle()
```

```python
class Prototype:
    """set global parameters"""
    B0 = 0.994
    R0 = 1.01
    phi_o0 = 0.80
    phi_y0 = 0.65

    #very rough approxs based on Japanese data
    annual_earning = 50000
    w0 = 30*annual_earning  #work for 30 years
    tax0 = 0.45
    b_y0 = 12*annual_earning
    T_o0 = 12*annual_earning
    b_o0 = 4*annual_earning
    
    
    def __init__(self, B = B0, R = R0, 
                 phi_o = phi_o0, phi_y = phi_y0, 
                 w = w0,  
                 b_y = b_y0, b_o = b_o0, 
                 tax = tax0, T_o = T_o0):
        self.B, self.R, self.phi_o, self.phi_y = B, R, phi_o, phi_y
        self.w, self.b_y, self.b_o, self.tax, self.T_o = w, b_y, b_o, tax, T_o
        
    def household(self):
        """Household Utility"""
        return (sp.log(self.Cy())) + self.B*self.phi_o*(sp.log(self.Co()))
    
    def S(self, R1 = R0, phi_y1 = phi_y0, phi_o1 = phi_o0, tax1 = tax0):
        """Household Saving"""
        B, R, phi_o, phi_y = self.B, self.R, self.phi_o, self.phi_y
        w, b_y, b_o, tax, T_o = self.w, self.b_y, self.b_o, self.tax, self.T_o
        return ((B * phi_o1/ (1 + B*phi_o1)) * (w * (1-tax1*phi_y1) + b_y/phi_y1)) + ((1 / (1+B*phi_o1)) * (b_o - T_o) / R1)
    
    def Cy(self, R1 = R0, phi_y1 = phi_y0, phi_o1 = phi_o0, tax1 = tax0):
        """Young Household Consumption"""
        B, R, phi_o, phi_y = self.B, self.R, self.phi_o, self.phi_y
        w, b_y, b_o, tax, T_o = self.w, self.b_y, self.b_o, self.tax, self.T_o
        return (1 / (1+B*phi_o1)) * (w * (1 - tax1*phi_y1) + b_y / phi_y1 + ((T_o - b_o)/R1))
    
    def Co(self, R1 = R0, phi_y1 = phi_y0, phi_o1 = phi_o0, tax1 = tax0):
        """Old Household Consumption"""
        B, R, phi_o, phi_y = self.B, self.R, self.phi_o, self.phi_y
        w, b_y, b_o, tax, T_o = self.w, self.b_y, self.b_o, self.tax, self.T_o
        return ((B*R1) / (1+B*phi_o1)) * (w* (1-tax1*phi_y1) + b_y/phi_y1) + (B / (1 + B*phi_o1)) * (T_o - b_o)
    
```

## <u>Firm</u>

A representativ Firm with Cobb-Douglas Technology (CRS): 

$$ F(K,L) = A K^\alpha L^{1-\alpha}$$

The Firm maximizes profit:

$$ \max_{K,L} F(K,L) - wL - rK $$

<u>F.O.Cs</u>: $$w = MP_L\quad \text{and}\quad r = MP_K$$

### <u>Capital Market Clearing</u>: 
$$K = sN$$

Thus, 

$$ K/N = \frac{\beta\phi_o}{1+\beta\phi_o}  \Big(A (1 - \tau_y \phi_y) (1-\alpha)(K/L)^\alpha  + \frac{b_y}{\phi_y}\Big) + \frac{1}{1+\beta \phi_o} \Big(\frac{b_o - T_o}{1 + A\alpha (K/L)^{(\alpha-1)} - \delta} \Big) $$

### <u>Government Budget Constraint</u>

\begin{align} N_o T_o + G &= \tau_y N_y w\\ \iff \phi_o N T_o + G &= \tau_y N w \quad (*) \end{align}



### <u>Economy-wide Resource constraint</u>

$$ C + S + G = Y + (1-\delta)K $$
where:
    
\begin{align}C &= N_yC_y + N_oC_o = NC_y + \phi_o N C_o &(*)\\ S &= N_y s = N s &(*) \\ G &= \tau_yN_yw - N_o T_o = \tau_y N w - \phi_o N T_o &(*) \end{align}

$(*)$ *assuming static population*


### <u>Law of motion of the economy</u>:

$$K^+ = (1-\delta)K + Y - C - G $$
